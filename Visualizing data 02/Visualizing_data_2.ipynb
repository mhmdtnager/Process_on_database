```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\mhmd\Downloads\Visualizing data 02\Orders.csv")
df
```

```python
Categories = df['Category'].value_counts()
Categories
```

```python
plt.plot(Categories , 'ro-')
```

```python
plt.plot(Categories , 'r--')
```

```python
plt.plot(Categories , 'r')
plt.xticks(rotation =45)
```

```python
plt.plot(Categories , 'go-')
plt.xticks(rotation =45)
plt.title('Orders by categories',fontsize=15)
plt.xlabel('Categories')
plt.ylabel('Orders')
plt.show()
```

```python
country_orders = df['Country'].value_counts()
country_orders
```

```python
plt.bar(country_orders.index ,country_orders.values)
plt.xticks(rotation = 90)
plt.title('Orders by countries',fontsize=15)
plt.xlabel('Country')
plt.ylabel('Orders')
plt.grid(axis='y', color='g', alpha=0.35, linestyle=':')#الخطوط بالعرض و اللون اخضر و الشفافية و يكون الخطوط منقطة
for ind, vl in enumerate(country_orders):
    plt.text(ind, vl, vl, ha='center', va='bottom')
```

```python
status = df['Status'].value_counts()
status
```

```python

plt.pie(status ,startangle=90 ,labels=status.values ,colors=['#2D3250','#F6B17A'])
plt.legend(labels=status.index , loc='upper left')
```

```python
my_labels = status.values/status.values.sum()*100
plt.pie(status ,startangle=90 ,labels=my_labels.round(1) ,colors=['#2D3250','#F6B17A'])
plt.legend(labels=status.index , loc='upper left')
```

```python
Cat_Quan_TotCost = df.groupby('Category')[['Quantity', 'Total Cost']].sum()
Cat_Quan_TotCost
```

```python
fig, axis = plt.subplots(nrows=2 , ncols=1)#هنا قولتلو يعملى صفين وعمود واحد على حسب انا عايز اعمل اي هكتب عدده
```

```python

fig, axis = plt.subplots(nrows=2, ncols=1)
fig.suptitle('Main Title')
fig.patch.set_facecolor('#e6e6e6')#الخلفية
fig.subplots_adjust(hspace=0.5)#عملت مسافة بين الرسمتين

#الرسمة الاولى
axis[0].plot(Cat_Quan_TotCost['Quantity'], 'r')
axis[0].set_xticks([])#شيلت الكلام من السينات
axis[0].set_title('Quantity by Category')

#الرسمة التانية
axis[1].plot(Cat_Quan_TotCost['Total Cost'])
axis[1].tick_params(axis='x', rotation=45)
axis[1].set_title('Cost by Category')
```
