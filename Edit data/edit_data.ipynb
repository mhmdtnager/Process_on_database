```python
df = pd.read_csv(r"C:\Users\mhmd\Downloads\Edit data\Orders.csv" , skiprows = 4)
df
```

```python
df.dropna(how = "all", inplace = True)#inplace عشان بدل معمل اسم جديد يكون نفس الاسم بالقيمة الجديدة
```

```python
df
```

```python
df['City'].str.title()#اول حرف كابتل
```

```python
df['City'] =df['City'].str.capitalize()
```

```python
df
```

```python
df['Branch']=df['Branch'].str.upper()
df['Email']=df['Email'].str.lower()
```

```python
df
```

```python
df['Phone Number']=df['Phone Number'].str.replace('Tel: ', '')#استبدلت كلمة tel بمكان فارغ عشان متبقاش موجودة
df
```

```python
df['Discount']=df['Discount'].round(1)#العلامة العشرية 
df
```

```python
df['Discount']=df['Discount'].round(0)
df
```

```python
df['Cost'] = df['Quantity'] * df['Unit Price']
```

```python
#انا مثلا عندى الكمية والسعر والتخفيض فعايز اعمل عملية تطلعلى الناتج دا فى عمود يعنى هضرب السعر ف الكمية واقسم على واطرح التخفيض
df['Total Price'] = df['Quantity'] * df['Unit Price'] - df['Discount'].fillna(0)#عوضت عن القيم الفارغة بصفر
df
```

```python

df.insert(9 , 'Customer Full Name' , df['Customer First Name']+' '+df['Customer Last Name'])#دمج عمودين مع بعض
```

```python
newcolumn = df['Branch'].str[:2]#هيجيب اول حرفين عشان كدا كئنه من الصفر إلى الواحد ، رقم اتنين مش معانا
df.insert(5 , 'Brcode' , newcolumn)
df
```

```python
df['Customer Full Name'] = df['Customer Full Name'].str.title()
pd.set_option('display.max_columns', None)#عشان لو فيه اعمدة مش ظاهرة عشان المكان تظهر (بيبقى مكانها نقط.....)
df
```

```python
df.loc[df['Status']== True]#كل الصح بس فى العمود دا
```

```python
df.loc[df['Status']== True , 'Status']
```

```python
#لو عايز اقول ان كل الترو دا تمم دوون (حاجة زى if)
df.loc[df['Status']== True , 'Order status'] = 'Done'
df.loc[df['Status']== False , 'Order status'] = 'Pending'
df.loc[(df['Status']== False) & (df['Order Date']<= '2023-01-01') , 'Order status'] = 'Canceled'

df
```

```python

def level(row):
    if row>1000:
        return 'High'
    else:
        return 'Low'
df['Cost Level'] = df['Cost'].apply(level) 
df
```

```python
import numpy as np
```

```python
df['Discount Descreption'] = np.where(df['Discount'].isnull(), 'No Discount', df['Discount']/df['Cost'] )
df
#np where 
#بتاخد تلت حجات الشرط الىى هو لو يساوى null التانى هنديله الاسم التالت لو لا يساوى null هنعمل المعادلة او نديله اسم الى احنا عايزينه
```
