```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\mhmd\Downloads\Visualizing data 03\Orders.csv")
df
```

```python
sns.histplot(df['Quantity'], bins=8)#bins عدد الاعمدة فى الجراف
```

```python
#محور اكس و محور واي و الداتا الى هيرسم منها و الالوان 

sns.barplot(x='Category' , y='Unit Price' , data=df , palette='crest')
plt.xticks(rotation = 90)
```

```python
#محور اكس و محور واي و الداتا الى هيرسم منها و الالوان 

sns.barplot(x='Category' , y='Unit Price' , data=df , palette='crest')
plt.xticks(rotation = 90)
plt.show()#عشان متظهرش اي حاجة غير الرسم
```

```python
#بشوف الخصم مناسب للتكلفة الكلية ولا لا وجبت الداتا و عملت حجم النقط على حسب الكمية لونت على حسب الصنف وعملت تدريج للون
sns.scatterplot(x='Discount' , y='Total Cost' , data=df ,size='Quantity' , hue='Category' , palette='crest' )
#المشكلة ان الجدول بتاع الالوان داخل فى الرسمة
```

```python
my_graph = sns.scatterplot(x='Discount' , y='Total Cost' , data=df ,size='Quantity' , hue='Category' , palette='crest' )
my_graph.legend(bbox_to_anchor=(1,1) , loc='upper left')
```

```python
#هنشوف كمية الطلبات على مدار الاسبوع هل فى ايام بيتطلب فيها اكتر مالتانية ولا لا
sns.boxplot(x='Day' , y='Quantity' , data=df , palette='crest')
```

```python
my_data = df.groupby(['Category' , 'Sub Category' , 'Item'])[['Quantity' , 'Total Cost' , 'Discount']].sum()
my_data
```

```python
#هحوله لجدول عادى من غير index
my_data = df.groupby(['Category' , 'Sub Category' , 'Item'], as_index=False)[['Quantity' , 'Total Cost' , 'Discount']].sum()
my_data
```

```python
sns.stripplot(x='Category' , y='Quantity' , data=my_data , palette='crest')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.jointplot(x='Quantity' , y='Total Cost' , data=my_data )
```

```python
sns.jointplot(x='Quantity' , y='Total Cost' , data=my_data , kind='kde' , fill=True )
```

```python
sns.pairplot(my_data)
```
