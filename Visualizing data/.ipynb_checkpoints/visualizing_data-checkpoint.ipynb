```python
import pandas as pd
```

```python
df= pd.read_csv(r"C:\Users\mhmd\Downloads\Visualizing data\Orders.csv")
df
```

```python
df.plot(x='Category', y='Total Cost')
```

```python
country_orders = df['Country'].value_counts()
country_orders
```

```python
country_orders.plot(kind = 'bar' , title = 'Orders By Country')
```

```python
cat_tot = df.groupby('Country')['Total Cost'].sum()
cat_tot
```

```python
cat_tot.plot(kind = 'barh', color = 'red' , xlabel = 'Total Cost(million)')
```

```python
cat_quan_tot = df.groupby('Category')[['Quantity', 'Total Cost']].sum()
cat_quan_tot
```
