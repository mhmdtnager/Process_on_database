```python
import pandas as pd
```

```python
df= pd.read_csv(r"C:\Users\mhmd\Downloads\Visualizing data\Orders.csv")
df
```

```python
df.plot(x='Category', y='Total Cost')
```

```python
country_orders = df['Country'].value_counts()
country_orders
```

```python
country_orders.plot(kind = 'bar' , title = 'Orders By Country')
```

```python
cat_tot = df.groupby('Country')['Total Cost'].sum()
cat_tot
```

```python
cat_tot.plot(kind = 'barh', color = 'red' , xlabel = 'Total Cost(million)')
```

```python
cat_quan_tot = df.groupby('Category')[['Quantity', 'Total Cost']].sum()
cat_quan_tot
```

```python
cat_quan_tot.plot()#النتيجة ان الكمية طالعة كدا لان فى فرق كبير بين الكميه والتكلفه كلها هنحل الموضوع دا فى الخلية الجاية
```

```python
cat_quan_tot.plot(subplots = True)
```

```python
#هنا كئنى عملت صادات على اليمين والشمال
tc_axis = cat_quan_tot.plot(y='Total Cost',kind = 'bar',color = 'red',ylabel = 'Total Cost (Million)')
cat_quan_tot.plot(y='Quantity',kind = 'line',secondary_y = True,ax=tc_axis, color='g', ylabel='Quantity',rot=45)
#axعشان بقله ان العمود الاصلى دا الى خد مكانه 
#rot45عشان الكلام الى تحت ميخشش فى بعضه فيميل شوية
```
